# Take Ckpts and Plot

```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import os
import sys
import pickle
import progressbar
```

# Oregon Slopebase and Oregon Offshore
- Whitened
- filtered between 20 and 120

```python
# Load Data for SlopeBase/Offshore with Whitening

bar = progressbar.ProgressBar(maxval=359, \
    widgets=[progressbar.Bar('=', '[', ']'), ' ', progressbar.Percentage()])
bar.start()

for k in range(153):   
    ckpt_name = os.path.dirname(os.path.dirname(os.getcwd())) + '/NCFs/ckpts_LJ01C_PC01A_0_152/ckpt_'+ str(k) +'.pkl'
    with open(ckpt_name, 'rb') as f:
        xcorr_1hr = pickle.load(f)
    
    if k == 0:
        xcorr = xcorr_1hr
    else:
        xcorr = xcorr + xcorr_1hr
    bar.update(k+1)
    
for k in range(153):   
    ckpt_name = os.path.dirname(os.path.dirname(os.getcwd())) + '/NCFs/ckpts_LJ01C_PC01A_153_306/ckpt_'+ str(k) +'.pkl'
    with open(ckpt_name, 'rb') as f:
        xcorr_1hr = pickle.load(f)

    xcorr = xcorr + xcorr_1hr
    bar.update(k+153)
    
for k in range(53):   
    ckpt_name = os.path.dirname(os.path.dirname(os.getcwd())) + '/NCFs/ckpts_LJ01C_PC01A_307_361/ckpt_'+ str(k) +'.pkl'
    with open(ckpt_name, 'rb') as f:
        xcorr_1hr = pickle.load(f)

    xcorr = xcorr + xcorr_1hr
    bar.update(k+153+153)
    
bar.finish()
```

```python
# Plot Xcorr
sns.set()

#Plot Xcorr
dt = 1/64000
t = np.arange(-xcorr.shape[0]*dt/2,xcorr.shape[0]*dt/2,dt)

fig = plt.figure(figsize=(7,5))
plt.plot(t,xcorr)
#plt.xlim([-30,30])
plt.title('NCF for 360 Hrs - PC01A/LJ01C',fontsize=16)
plt.xlabel('Delay Time Ï„ (s)')

fig.savefig('NCF_360hrs_whiten.png',dpi=500)
```
